```python
def serialization(u,seq,len_seq):
    length=len(seq)
    ser_list=[]
    for i in range(length-len_seq):
        #tmp = np.zeros([len_seq], dtype=np.int32)
        #tmp=seq[:len_seq+i]
        tmp=seq[i:len_seq+i]
        nxt=seq[len_seq+i]
        ser_list.append((u,list(tmp),nxt))
    return ser_list
```

```python
import random

def non_zero_sample(user_dict,maxlen,len_seq,del_u,del_seq_len):
    train_set=[]
    test_set=[]
    val_set=[]
    train_val_set=[]
    for u in user_dict.keys():
        seq=user_dict[u]
        length=len(seq)
        if length <len_seq+3:
            del_seq_len.append(length)
            del_u.append(length)
            print(u,length)
            continue
        len_test=max(1,int(0.2*(length-len_seq)))
        len_val=max(1,int(0.1*(length-len_seq)))
        test_seq=seq[-len_test-len_seq:]
        val_seq=seq[-len_test-len_val-len_seq:-len_test]
        train_seq=seq[:-len_test-len_val]

        #print(length,len(test_seq),len(val_seq),len(train_seq))
        if len(test_seq)<=10|len(val_seq)<=10|len(train_seq)<=10:
            print('wrong!')
        for i in serialization(u,train_seq,len_seq):
            train_set.append(i)
        for i in serialization(u,test_seq,len_seq):
            test_set.append(i)
        for i in serialization(u,val_seq,len_seq):
            val_set.append(i)

    random.shuffle(train_set)
    random.shuffle(test_set)
    return train_set,test_set ,val_set
```

```python
del_u=[]
del_seq_len=[]
train_set,test_set,val_set=non_zero_sample(k100,509,len_seq,del_u,del_seq_len)
print(len(train_set),len(test_set),len(val_set))
if len(train_set)+len(test_set)+len(val_set)+sum(del_seq_len)+(users_num-len(del_u))*len_seq==sum(len(i) for i in k100.values()):
    print('Preprocess Perfect')
```

```python
users, click_seqs, pos_items = zip(*train_set)
train_data = {'users':np.array(users),'click_seq': np.array(click_seqs), 'pos_item': np.array(pos_items)}

users, click_seqs, pos_items= zip(*test_set)
test_data = {'users':np.array(users),'click_seq': np.array(click_seqs), 'pos_item': np.array(pos_items)}

users, click_seqs, pos_items = zip(*val_set)
val_data = {'users':np.array(users),'click_seq': np.array(click_seqs), 'pos_item': np.array(pos_items)}
```
